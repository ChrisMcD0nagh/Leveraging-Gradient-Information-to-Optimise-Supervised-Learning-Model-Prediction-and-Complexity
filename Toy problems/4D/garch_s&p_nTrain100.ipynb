## Define Garch(1,1) model

```python
nInputDim = 4 # Latent parameters
nOutputDim = 1 # Vol output
T = 5 # Future window range
forecast_window = range(1, T + 1)
```

```python
# GARCH (1,1) process that only returns conditional stds
def garch(params, conditional, ret, t, prev_sigma=None):
    
    # Explicity define parameters
    mu = params[0]
    omega = params[1]
    alpha = params[2]
    beta = params[3]
    
    resid = ret - mu
    
    if t == 1:
        sigma2 = omega + alpha * resid**2 + beta * conditional ** 2
    else: 
        sigma2 = omega + (alpha + beta) * prev_sigma ** 2
        
    return sigma2 ** 0.5
```

## Estimating paramters with Maximum likelihood

```python
def ll_get_pred(curr_surrogate, theta, return_series):
    mu = theta[0]
    omega = theta[1]
    alpha = theta[2]
    beta = theta[3]
    
    lr =  (omega/(1 - alpha - beta))**0.5
    T = len(return_series)
    synPred = np.zeros(T)
    synPred[0] = lr
    for t in range(1, T):
        synPred[t] = curr_surrogate[1].predict_on_batch([theta.reshape((1, nInputDim)), synPred[t-1].reshape((1,1)), np.array(return_series.iloc[t-1]).reshape((1,1))])[0].flatten()[0]
    return synPred
```

```python
def garch_mle_concurrent(theta, return_series, surrogate_models):

    return_series = return_series[-200:]
    mu = theta[0]
    omega = theta[1]
    alpha = theta[2]
    beta = theta[3]
    
    lr =  (omega/(1 - alpha - beta))**0.5

    resid = return_series - mu
    realised = abs(resid)

    T = len(return_series)
    synPred_avg = np.zeros(T)

    with concurrent.futures.ThreadPoolExecutor(max_workers=no_models) as executor:
        future_to_res = {executor.submit(ll_get_pred, surrogate_models[m], theta, return_series): m \
                         for m in range(no_models)}
        for i in concurrent.futures.as_completed(future_to_res):
            synPred = i.result()
            synPred_avg += synPred
    
    synPred_avg /= no_models
    # Compute likelihood of predictions
    log_likelihood = sum(-np.log(synPred_avg * (2 * np.pi) ** 0.5) - (realised ** 2 / (2 * synPred_avg **2)))
    #print(log_likelihood)
    return -log_likelihood
```

```python
def garch_mle(theta, return_series, surrogate_models, thresh=200):

    return_series = return_series[-thresh:]
    mu = theta[0]
    omega = theta[1]
    alpha = theta[2]
    beta = theta[3]
    
    lr =  (omega/(1 - alpha - beta))**0.5

    resid = return_series - mu
    realised = abs(resid)

    T = len(return_series)
    synPred_avg = np.zeros(T)
    synPred_avg[0] = lr
    
    for m in range(no_models):
        synPred = np.zeros(T)
        synPred[0] = lr
        curr_surrogate = surrogate_models[m]
        for t in range(1, T):
            synPred[t] = curr_surrogate[1].predict_on_batch([theta.reshape((1, nInputDim)), synPred[t-1].reshape((1,1)), np.array(return_series.iloc[t-1]).reshape((1,1))])[0].flatten()[0]
        synPred_avg += synPred
    synPred_avg /= no_models
    # Compute likelihood of predictions
    log_likelihood = sum(-np.log(synPred_avg * (2 * np.pi) ** 0.5) - (realised ** 2 / (2 * synPred_avg **2)))
    return -log_likelihood
```

```python
def garch_mle_TRUE(theta, return_series, thresh=200):

    return_series = return_series[-thresh:].values
    mu = theta[0]
    omega = theta[1]
    alpha = theta[2]
    beta = theta[3]
    
    lr =  (omega/(1 - alpha - beta))**0.5

    resid = return_series - mu
    realised = abs(resid)

    T = len(return_series)
    synPred = np.zeros(T)
    synPred[0] = lr

    for t in range(1, T):
        #synPred[t] = (omega + alpha*resid[t-1]**2 + beta*synPred[t-1]**2)**(1/2)
        synPred[t] = garch(theta, synPred[t-1], return_series[t-1], 1)
    #print(f'loop completed: {time()-t0:.02f}s')
    # Compute likelihood of predictions
    log_likelihood = sum(-np.log(synPred * (2 * np.pi) ** 0.5) - (realised ** 2 / (2 * synPred **2)))
    return -log_likelihood
```

```python
%%time

x0 = np.array([train_series.mean(), train_series.std()**2, 0.5, 0.5])

with concurrent.futures.ThreadPoolExecutor(max_workers=3) as executor:
        f1_std = executor.submit(spop.minimize, garch_mle, x0, args=(train_series, surrogate_std, split), method='Nelder-Mead', tol=1e-3)
        f2_grad = executor.submit(spop.minimize, garch_mle, x0, args=(train_series, surrogate_grad, split), method='Nelder-Mead', tol=1e-3)
        f3_true = executor.submit(spop.minimize, garch_mle_TRUE, x0, args=(train_series, split), method='Nelder-Mead')

params_std = f1_std.result().x
params_grad = f2_grad.result().x
params_func = f3_true.result().x
```

```python
fig=plt.figure(figsize=(10,5), facecolor='white')
plt.rcParams.update({'font.size': 10})
ax1 = plt.subplot(111)

ind = np.arange(nInputDim)

std_diff_MLE = abs(params_std - params_func)
grad_diff_MLE = abs(params_grad - params_func)

width = 0.2      

rects1 = ax1.bar(ind, std_diff_MLE, width, color='b', label='Std surrogate difference')
rects2 = ax1.bar(ind+width, grad_diff_MLE, width, color='g', label='Grad surrogate difference')

ax1.set_ylabel(r'Absolute Difference')
ax1.set_xticks(ind + width / 2)
ax1.set_xticklabels( (r'$\mu$', r'$\omega$', r'$\alpha$', r'$\beta$') )
ax1.set_xlabel('Parameter')
plt.legend();
#plt.savefig('GARCH_mle_diff_nTrain100.png', bbox_inches='tight')
```

```python
def pred_vol(theta, return_series, surrogate=None, use_func = False):
    T = len(return_series)

    mu = theta[0]
    omega = theta[1]
    alpha = theta[2]
    beta = theta[3]
    
    lr =  (omega/(1 - alpha - beta))**0.5
    synPred_avg = np.zeros(T)
    if use_func:
        synPred_avg[0] = lr
        for t in range(1, T):
            synPred_avg[t] = garch(theta, synPred_avg[t-1], return_series[t-1], 1)
    else:
        for m in range(no_models):
            synPred = np.zeros(T)
            curr_model = surrogate[m]
            for t in range(1, T):
                synPred[t] = curr_model[1].predict_on_batch([theta.reshape((1, nInputDim)), synPred[t-1].reshape((1,1)), np.array(return_series.iloc[t-1]).reshape((1,1))])[0].flatten()[0]
            synPred_avg += synPred
        synPred_avg /= no_models
    return synPred_avg
```

```python
fig=plt.figure(figsize=(15,8), facecolor='white')
plt.rcParams.update({'font.size': 10})
ax1 = plt.subplot(111)

realised_vol = abs(train_series - params_func[0])
pred_std = pred_vol(params_std, train_series, surrogate_std)
pred_grad = pred_vol(params_grad, train_series, surrogate_grad)
pred_func = pred_vol(params_func, train_series, use_func=True)

ax1.plot(prices.index[1:split+1], realised_vol, label='Realised volatility', c='orange', linestyle=':', alpha=0.8)
ax1.plot(prices.index[1:split+1], pred_std, label='Standard surrogate', c='b')
ax1.plot(prices.index[1:split+1], pred_grad, label='Gradient surrogate', c='g')
ax1.plot(prices.index[1:split+1], pred_func, linestyle='--', alpha=0.5, label='True function', c='r')
ax1.set_ylabel('Volatility')
ax1.set_xlabel('Date')
ax1.legend()
plt.savefig('sp500_train_pred_comp._nTrain100.png', bbox_inches='tight')
```

```python
def rmse(pred, true):
    return (sum((pred - true)**2) /len(pred))**0.5
```

```python
std_rmse = rmse(realised_vol, pred_std)
grad_rmse = rmse(realised_vol, pred_grad)
func_rmse = rmse(realised_vol, pred_func)
print(f'Training RMSE:')
print(f' Grad surrogate: {grad_rmse:.04f}')
print(f' Std surrogate: {std_rmse:.04f}')
print(f' True function: {std_rmse:.04f}')
print(f' Volatility std. dev: {realised_vol.std():.04f}')
```

## Test prediction with rolling training series

```python
def rolling_pred(return_series, init_params, surrogate=None, use_func=False, interval=50):
    
    # Empty DF to store predictions
    rolling_conditional_vol_preds = pd.DataFrame(index=np.arange(test_size), columns=np.arange(test_size))
    rolling_conditional_vol_preds.loc[0, 0] = (init_params[1]/(1 - init_params[2] - init_params[3]))**0.5
    
    # Store rolling standard deviation to provide baseline performance
    rolling_std_pred = []
    params = init_params
    for i in tqdm(range(0, test_size)):
        train = pd.Series(return_series[:-(test_size-i)])
        if i % interval == 0 and i != 0 and use_func == False:
            # Update parameters using last 200 days of training data
            res = spop.minimize(garch_mle_TRUE, params, args=(train), method='Nelder-Mead')
            params = res.x

        rolling_std_pred.append(train.std())

        ret = train.iloc[-1]

        min_i = max(0, i-1)
        for t in range(1, 5 + 1):
            if i + t > test_size:
                break
            cond1 = rolling_conditional_vol_preds.loc[min_i, i] # For t+1 use prior t+1 prediction
            cond2 = rolling_conditional_vol_preds.loc[i, i+t-1] # Use expectation relationship for t + h h >= 2
            if t == 1:
                if use_func:
                    rolling_conditional_vol_preds.loc[i, i+t] = garch(params, cond1, ret, t, prev_sigma=cond2)
                else:
                    pred = 0
                    for m in range(no_models):
                        curr_surrogate = surrogate[m]
                        pred += curr_surrogate[t].predict([params.reshape((1, nInputDim)), np.array(cond1).reshape((1,1)), 
                                                                                  np.array((ret)).reshape((1,1))])[0].flatten()[0]
                    pred /= no_models
                    rolling_conditional_vol_preds.loc[i, i+t] = pred
                                                                        
            else:
                if use_func:
                    rolling_conditional_vol_preds.loc[i, i+t] = garch(params, cond1, ret, t, prev_sigma=cond2)
                else:
                    pred = 0
                    for m in range(no_models):
                        curr_surrogate = surrogate[m]
                        pred += curr_surrogate[t].predict([params.reshape((1, nInputDim)), np.array(cond2).reshape((1,1))])[0].flatten()[0]
                    pred /= no_models
                    rolling_conditional_vol_preds.loc[i, i+t] = pred

    
    return rolling_conditional_vol_preds, rolling_std_pred
    
```

```python
%%time

MLE_int = 1 # How often parameters are updated with MLE on new data
with concurrent.futures.ThreadPoolExecutor(max_workers=3) as executor:
        f1_std = executor.submit(rolling_pred, returns, params_func, surrogate_std, interval=MLE_int)
        f2_grad = executor.submit(rolling_pred, returns, params_func, surrogate_grad, interval=MLE_int)
        f3_func = executor.submit(rolling_pred, returns, params_func, use_func=True, interval=MLE_int)
rolling_pred_std, std_baseline = f1_std.result()
rolling_pred_grad, _ = f2_grad.result()
rolling_pred_func, _ = f3_func.result()
```
