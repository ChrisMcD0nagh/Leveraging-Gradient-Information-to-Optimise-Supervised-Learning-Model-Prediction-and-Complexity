```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from os import listdir
from collections import defaultdict
```

```python
# Load in experimental results
foldernames = ['2D', '4D', '8D']
all_results = defaultdict(dict)
base = '../Experimental results'
for ts_folder in foldernames:
    files = listdir(base+'/'+ts_folder)
    for f in files:
        if not f.startswith('.'):
            all_results[ts_folder][f] = {}
            for r in listdir(f'{base}/{ts_folder}/{f}'):
                if 'grad_test_corr' in r:
                    all_results[ts_folder][f]['grad_test_corr'] = pd.read_csv(f'{base}/{ts_folder}/{f}/{r}', index_col=0)
                elif 'grad_train_corr' in r:
                    all_results[ts_folder][f]['grad_train_corr'] = pd.read_csv(f'{base}/{ts_folder}/{f}/{r}', index_col=0)
                elif 'grad_test_rmse' in r:
                    all_results[ts_folder][f]['grad_test_rmse'] = pd.read_csv(f'{base}/{ts_folder}/{f}/{r}', index_col=0)
                elif 'grad_train_rmse' in r:
                    all_results[ts_folder][f]['grad_train_rmse'] = pd.read_csv(f'{base}/{ts_folder}/{f}/{r}', index_col=0)
                elif 'std_train_corr' in r:
                    all_results[ts_folder][f]['std_train_corr'] = pd.read_csv(f'{base}/{ts_folder}/{f}/{r}', index_col=0)
                elif 'std_test_rmse' in r:
                    all_results[ts_folder][f]['std_test_rmse'] = pd.read_csv(f'{base}/{ts_folder}/{f}/{r}', index_col=0)
                elif 'std_train_rmse' in r:
                    all_results[ts_folder][f]['std_train_rmse'] = pd.read_csv(f'{base}/{ts_folder}/{f}/{r}', index_col=0)
                elif 'std_test_corr' in r:
                    all_results[ts_folder][f]['std_test_corr'] = pd.read_csv(f'{base}/{ts_folder}/{f}/{r}', index_col=0)
```

## 2D problem

### Model accuracy evaluation

```python
complexity_flag = '2D'
fig=plt.figure(figsize=(20,15), facecolor='white')
plt.rcParams.update({'font.size': 10})
ax1 = plt.subplot(221)
ax2 = plt.subplot(222)
ax3 = plt.subplot(223)
ax4 = plt.subplot(224)

lw = 2
min_c = 2
max_c = 60
interval = 4


axes = [ax1, ax2, ax3, ax4]
linestyles = ['--', '-', ':', '-.', (0, (3, 5, 1, 5))]
markers = ['s', 'D', 'v', 'p', '*']
std_colours = ['#FFBE0B', '#EE432F', '#39AFF9', '#79B995', '#9B3BCE']
grad_colours = ['#B88700', '#95190C', '#04578B', '#417B5A', '#6E2594']
diff_colours = ['#E0A500', '#E22712', '#0795ED', '#5FAB80', '#842DB4']
samples = ['nTrain50', 'nTrain100', 'nTrain300', 'nTrain500', 'nTrain700']

for s, ls, m, std_c, grad_c, diff_c in zip(samples, linestyles, markers, std_colours, grad_colours, diff_colours):
    all_results[complexity_flag][s]['grad_test_rmse']['mean'].plot(ax=ax1, label=f'{s} + GRAD', c=grad_c, linestyle=ls, linewidth=lw, marker=m);
    all_results[complexity_flag][s]['std_test_rmse']['mean'].plot(ax=ax1, label=f'{s}', c=std_c, linestyle=ls, linewidth=lw, marker=m);
    
    rmse_diff = pd.DataFrame(index=all_results[complexity_flag][s]['grad_test_rmse'].index)
    rmse_diff[s] = (all_results[complexity_flag][s]['std_test_rmse']['mean'] - all_results[complexity_flag][s]['grad_test_rmse']['mean'] \
                   ) * 100 / all_results[complexity_flag][s]['std_test_rmse']['mean']
    rmse_diff.plot(ax=ax2, linestyle=ls, c=diff_c, lw=lw, legend=False, marker=m)
    
    
    corr_diff = pd.DataFrame(index=all_results[complexity_flag][s]['grad_test_corr'].index)
    corr_diff[s] = (all_results[complexity_flag][s]['grad_test_corr']['mean'] - all_results[complexity_flag][s]['std_test_corr']['mean'] \
                   ) * 100 / all_results[complexity_flag][s]['std_test_corr']['mean']
    corr_diff.plot(ax=ax4, linestyle=ls, c=diff_c, lw=lw, legend=False, marker=m)
    
    
    all_results[complexity_flag][s]['grad_test_corr']['mean'].plot(ax=ax3, label=f'{s} + GRAD', c=grad_c, linestyle=ls, linewidth=lw, marker=m);
    all_results[complexity_flag][s]['std_test_corr']['mean'].plot(ax=ax3, label=f'{s}', c=std_c, linestyle=ls, linewidth=lw, marker=m);

ax2.axhline(0, linestyle='-', alpha=1, c='#0D080B', linewidth=1)    
ax4.axhline(0, linestyle='-', alpha=1, c='#0D080B', linewidth=1)

ax1.set_ylabel('RMSE')
ax3.set_xlabel(r'Model complexity $\eta$')
ax4.set_xlabel(r'Model complexity $\eta$')
ax2.set_ylabel('Difference (%)')
ax3.set_ylabel('Perason correlation')
ax4.set_ylabel('Difference (%)');
ax1.set_xticks(range(min_c, max_c)[::interval])
ax2.set_xticks(range(min_c, max_c)[::interval])
ax3.set_xticks(range(min_c, max_c)[::interval])
ax4.set_xticks(range(min_c, max_c)[::interval])
handles1, labels1 = ax1.get_legend_handles_labels()
handles2, labels2 = ax2.get_legend_handles_labels()
fig.legend(handles1, labels1, loc='center left')
fig.legend(handles2, labels2, loc='center right')
ax1.grid()
ax2.grid()
ax3.grid()
ax4.grid()
```

### Overfitting analysis

```python
complexity_flag = '2D'

fig=plt.figure(figsize=(25,20), facecolor='white')
plt.rcParams.update({'font.size': 15})
ax1 = plt.subplot(321)
ax2 = plt.subplot(322)
ax3 = plt.subplot(323)
ax4 = plt.subplot(324)
ax5 = plt.subplot(325)

lw = 3
axes = [ax1, ax2, ax3, ax4, ax5]
samples = ['nTrain50', 'nTrain100', 'nTrain300', 'nTrain500', 'nTrain700']

for s, ax in zip(samples, axes):
    all_results[complexity_flag][s]['grad_test_rmse']['mean'].plot(ax=ax, label='Test (Grad)', c='g', linestyle='-', lw=lw)
    all_results[complexity_flag][s]['std_test_rmse']['mean'].plot(ax=ax, label='Test (Std)', c='r', linestyle='-', lw=lw)
    all_results[complexity_flag][s]['grad_train_rmse']['mean'].plot(ax=ax, label='Train (Grad)', c='g', linestyle=':', lw=lw)
    all_results[complexity_flag][s]['std_train_rmse']['mean'].plot(ax=ax, label='Train (Std)', c='r', linestyle=':', lw=lw)
    
    ax.set_ylabel('RMSE')
    ax.set_xlabel('Model complexity')
    ax.set_title(s)
    ax.legend();
    
ax1.set_ylim([0,  0.05]);
ax2.set_ylim([0,  0.03]);
ax3.set_ylim([0,  0.01]);
ax4.set_ylim([0,  0.006]);
ax5.set_ylim([0,  0.006]);
```
